```python
import pandas as pd
import numpy as np
```

```python
filename = '../data/geonames_countryInfo.txt'
meta1 = pd.read_csv(filename, skiprows=49, usecols=range(11), keep_default_na=False)
isos1 = set(pd.unique(meta1['ISO']))
country1 = set(pd.unique(meta1['Country']))

print(meta1.shape)
meta1.head()
```

```python
filename = '../data/UNSD_Methodology.csv'
meta2 = pd.read_csv(filename, sep=';', keep_default_na=False)
meta2 = meta2.loc[:, ['ISO','M49 Code','Region Code','Region Name','Subregion Code','Subregion Name']]
isos2 = set(pd.unique(meta2['ISO']))

print(meta2.shape)
meta2.head()
```

```python
filename = '../data/WUP2018-F01-Total_Urban_Rural.xls'
meta3 = pd.read_excel(filename, skiprows=16)
meta3 = meta3.rename(columns={'Country\ncode':'M49 Code'}).drop(columns=['Index', 'Note'])

print(meta3.shape)
meta3.head()
```

```python
meta = meta1.merge(meta2.merge(meta3, on='M49 Code', how='inner'), how='inner', on='ISO')
meta = meta[meta['Region Name'] != '']
print(meta.shape)
meta.head()
```

```python
meta['Continent'] = meta['Subregion Name']

meta.loc[meta['Region Name'] == 'Europe', 'Continent'] = 'Europe'
meta.loc[meta['Region Name'] == 'Oceania', 'Continent'] = 'Oceania'

meta.loc[meta['Subregion Name'] == 'Western Asia', 'Continent'] = 'WANA'
meta.loc[meta['Subregion Name'] == 'Northern Africa', 'Continent'] = 'WANA'

meta.loc[meta['Subregion Name'] == 'Southern Asia', 'Continent'] = 'Rest of Asia'
meta.loc[meta['Subregion Name'] == 'South-eastern Asia', 'Continent'] = 'Rest of Asia'
meta.loc[meta['Subregion Name'] == 'Eastern Asia', 'Continent'] = 'Rest of Asia'
meta.loc[meta['Subregion Name'] == 'Central Asia', 'Continent'] = 'Rest of Asia'

meta.loc[meta['Country'] == 'China', 'Continent'] = 'China'
meta.loc[meta['Country'] == 'Hong Kong', 'Continent'] = 'China'
meta.loc[meta['Country'] == 'Macao', 'Continent'] = 'China'
```

```python
if 1 == 0:
    meta.loc[meta['Region Name'] == 'Europe', 'Subregion Name'] = 'Europe'
    meta.loc[meta['Region Name'] == 'Oceania', 'Subregion Name'] = 'Oceania'

    meta.loc[meta['Subregion Name'] == 'Western Asia', 'Subregion Name'] = 'WANA'
    meta.loc[meta['Subregion Name'] == 'Northern Africa', 'Subregion Name'] = 'WANA'

    meta.loc[meta['Subregion Name'] == 'Southern Asia', 'Subregion Name'] = 'Asia'
    meta.loc[meta['Subregion Name'] == 'South-eastern Asia', 'Subregion Name'] = 'Asia'
    meta.loc[meta['Subregion Name'] == 'Eastern Asia', 'Subregion Name'] = 'Asia'
    meta.loc[meta['Subregion Name'] == 'Central Asia', 'Subregion Name'] = 'Asia'

    meta.loc[meta['Country'] == 'China', 'Subregion Name'] = 'China'
    meta.loc[meta['Country'] == 'Hong Kong', 'Subregion Name'] = 'China'
    meta.loc[meta['Country'] == 'Macao', 'Subregion Name'] = 'China'

    meta.loc[meta['Subregion Name'] == 'Melanesia', 'Subregion Name'] = 'Other'
    meta.loc[meta['Subregion Name'] == 'Polynesia', 'Subregion Name'] = 'Other'
    meta.loc[meta['Subregion Name'] == 'Micronesia', 'Subregion Name'] = 'Other'
```

```python
filename = '../data/geonames_UNSD_WUP_metadata.csv'
meta.to_csv(filename, index=False)
```

```python
field = 'Income'
field = 'Continent'
incomes = sorted(pd.unique(meta[field]))
print(incomes)
urbpop = np.zeros(len(incomes))
rurpop = np.zeros(len(incomes))
totpop = np.zeros(len(incomes))

total = np.sum(meta['Total'])

for i in range(len(incomes)):
    df = meta[meta[field] == incomes[i]]
    urbpop[i] = np.sum(df['Urban'])
    rurpop[i] = np.sum(df['Rural'])
    totpop[i] = np.sum(df['Total'])
    
toturb = np.sum(urbpop)
totrur = np.sum(rurpop)
```

```python
df = pd.DataFrame()
df['Urban'] = urbpop/toturb*100
df['Rural'] = rurpop/totrur*100
df['Total'] = totpop/total*100
df.index = incomes
df
```

```python
filename = '../results/' + field.lower().replace(' ','_') + '_population.csv'
print(filename)
df.to_csv(filename, index=False)
```
